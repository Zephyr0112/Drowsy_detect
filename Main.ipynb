```python
#!.\dd\Scripts\activate

import cv2
import os
from keras.models import load_model
import numpy as np
from pygame import mixer
import time


mixer.init()
sound = mixer.Sound('alarm.wav')

face = cv2.CascadeClassifier('haar cascade files\haarcascade_frontalface_alt.xml')
leye = cv2.CascadeClassifier('haar cascade files\haarcascade_lefteye_2splits.xml')
reye = cv2.CascadeClassifier('haar cascade files\haarcascade_righteye_2splits.xml')



lbl=['Close','Open']

model = load_model('models/mod_95.h5')
path = os.getcwd()
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
count=0
score=0

rpred=[99]
lpred=[99]

while(True):
    ret, frame = cap.read()
    height,width = frame.shape[:2] 

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = face.detectMultiScale(gray,minNeighbors=5,scaleFactor=1.1,minSize=(25,25))
    left_eye = leye.detectMultiScale(gray)
    right_eye =  reye.detectMultiScale(gray)

    cv2.rectangle(frame, (0,height-50) , (200,height) , (0,0,0) , thickness=cv2.FILLED )

    for (x,y,w,h) in right_eye:
        cv2.rectangle(frame, (x,y) , (x+w,y+h) , (100,100,100) , 1 )
        r_eye=frame[y:y+h,x:x+w]
        count=count+1
        r_eye = cv2.cvtColor(r_eye,cv2.COLOR_BGR2GRAY)
        r_eye = cv2.resize(r_eye,(80,80))
        r_eye= r_eye/255
        r_eye=  r_eye.reshape(80,80,-1)
        r_eye = np.expand_dims(r_eye,axis=0)

        rpred = model.predict(r_eye, verbose = 0)[0][0]
        print(rpred)
        if(rpred<0.5):
            lbl='Open' 
        else:
            lbl='Closed'
        break

    for (x,y,w,h) in left_eye:
        cv2.rectangle(frame, (x,y) , (x+w,y+h) , (100,100,100) , 1 )
        l_eye=frame[y:y+h,x:x+w]
        count=count+1
        l_eye = cv2.cvtColor(l_eye,cv2.COLOR_BGR2GRAY)  
        l_eye = cv2.resize(l_eye,(80,80))
        l_eye= l_eye/255
        l_eye=l_eye.reshape(80,80,-1)
        l_eye = np.expand_dims(l_eye,axis=0)

        lpred = model.predict(l_eye, verbose = 0)[0][0]
        print(lpred)
        if(lpred<0.3):
            lbl='Open'   
        else:
            lbl='Closed'
        break
    
    print(rpred, lpred)
    
    if(rpred>0.5 and lpred>0.5):
        score=score+1
        
    # if(rpred[0]==1 or lpred[0]==1):
    else:
        score=score-1
        
    
        
    if(score<0):
        score=0   
    cv2.putText(frame,'Current:'+str(score),(0,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    if(score>2):
        #person is feeling sleepy so we beep the alarm
        cv2.putText(frame,"ALERT, DROWSY",(width-200,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
        cv2.imwrite(os.path.join(path,'image.jpg'),frame)
        try:
            sound.play()
            
        except:  # isplaying = False
            pass
     
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
